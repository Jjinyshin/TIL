````python
# 매장 정보 리스트 만들기
starbucks_list = []
# 시작시간
start = time.time()
print('[스크래핑 시작...]')

# 스크래핑 과정
# for btn in range(len(next_btn))[1:]:  # '<' 버튼 무시
# 페이지 안넘어감 -> while문으로 변경
while True:
  # frame 변경
  switch_frame('searchIframe')
  page_down(40)
  sleep(3)

  # 스벅 리스트
  sbuck_list = driver.find_elements(By.CSS_SELECTOR, 'li.UEzoS')
  # 스벅 매장이름
  names = driver.find_elements(By.CSS_SELECTOR, '.TYaxT')
  # 매장 주소
  addr = driver.find_elements(By.CSS_SELECTOR, '.hClKF')
  # 페이지 리스트(다음 페이지)
  next_btn = driver.find_elements(By.CSS_SELECTOR, '.eUTV2')
  # 별점, 리뷰 span의 class h69bs
  reviews = driver.find_elements(By.CSS_SELECTOR, '.h69bs')
  # 리뷰 구분을 위한 인덱스 변수
  rev_idx = 1
  # 별점
  rating = np.nan

  for sbuck in range(len(sbuck_list)):
    print(sbuck, end=" ")
    sleep(1)

    try:
      # 스타벅스 매장이름 가져오기
      sbuck_name = names[sbuck].text
      print(sbuck_name, end=" ")

      # 주소 가져오기 - 서울 00구 XX동 형식
      address = addr[sbuck].text
      print(address, end=" ")

      # 별점, 리뷰 가져오기
      if rev_idx >= (len(reviews) - 1): break
      visitor_reviews = reviews.__getitem__(rev_idx).text
      rev_idx += 1
      blog_reviews = reviews.__getitem__(rev_idx).text
      if len(reviews.__getitem__(rev_idx + 2).get_attribute("class").split(" ")) == 2:
        # '별점' 이라면
        rating = reviews.__getitem__(rev_idx+2).text.split("\n")[1]
        print(rating, end=" ")
        rev_idx += 3
      else: rev_idx += 2

      print(visitor_reviews, end=" ")
      print(blog_reviews)

      dict_temp = {
      'Name': sbuck_name,
      'Address': address,
      'VisitorReviews': visitor_reviews,
      'BlogReviews': blog_reviews,
      'Rating': rating
      }

      starbucks_list.append(dict_temp)

    except Exception as e:
      print(e)
      print('ERROR!' * 3)
  
  # 마지막 페이지일때: '>'버튼의 속성aria-disabled="true" 이면 종료
  if next_btn[-1].get_attribute('aria-disabled') == 'true':
    print('드디어 끝!!!')
    break

  if names[-1]:  # 마지막 매장일 경우 다음버튼 클릭
      next_btn[-1].click()
      sleep(2)

  else:
      print('페이지 인식 못함')
      break

print('[데이터 수집 완료]\n소요 시간 :', time.time() - start)
# 작업 종료 후 창닫기
driver.quit()
````
